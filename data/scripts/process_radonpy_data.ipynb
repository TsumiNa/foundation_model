```python
import joblib
import pandas as pd
from sklearn.discriminant_analysis import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PowerTransformer
```

```python
radonpy_data = pd.read_csv("../amorphous_polymer_all_data_FFDescriptor_20250730.csv", index_col=0)
radonpy_data.head()
```

```python
FF_desc_names = radonpy_data.columns.tolist()[-190:]
FF_desc = radonpy_data[FF_desc_names]
FF_desc.to_parquet("../amorphous_polymer_FFDescriptor_20250730.parquet")
FF_desc
```

```python
prop_names_with_stats = radonpy_data.columns.tolist()[:-214]
print(len(prop_names_with_stats))
prop_names = [p for p in prop_names_with_stats if not p.endswith(("_median", "_std", "_min", "_max", "_count"))]
print(len(prop_names))

props = radonpy_data[prop_names]
# Create a dictionary to store pipelines for each property
pipes = {}
scaled_props_list = []

# Process each property separately
for col in props.columns:
    pipe = Pipeline(
        [("standardscaler", StandardScaler()), ("powertransformer", PowerTransformer(method="yeo-johnson"))]
    )
    # Fit and transform single column data
    data = props[[col]].values
    scaled_data = pipe.fit_transform(data)
    # Store transformed data
    scaled_props_list.append(pd.DataFrame(data=scaled_data, index=props.index, columns=[f"{col} (normalized)"]))
    # Store pipeline
    pipes[col] = pipe
# Save pipeline
joblib.dump(pipes, "../amorphous_polymer_properties_scaler_20250730.pkl.z")

# Combine all scaled properties
scaled_props = pd.concat(scaled_props_list, axis=1)

props = pd.concat([props, scaled_props], axis=1)
```

```python
from sklearn.model_selection import train_test_split

train_idx, val_idx, test_idx = [], [], []

for prefix in ["VL", "PI", "PG"]:
    data = props[props.index.str.startswith(prefix)]
    idx = data.index.tolist()
    n = len(idx)
    if n < 5:
        train_idx.extend(idx)
        continue
    idx_train, idx_temp = train_test_split(idx, test_size=0.3, random_state=42)
    idx_val, idx_test = train_test_split(idx_temp, test_size=0.5, random_state=42)
    train_idx.extend(idx_train)
    val_idx.extend(idx_val)
    test_idx.extend(idx_test)

split = pd.Series(index=props.index, dtype="object")
split.loc[train_idx] = "train"
split.loc[val_idx] = "val"
split.loc[test_idx] = "test"
props["split"] = split
```

```python
props.to_parquet("../amorphous_polymer_properties_20250730.parquet")
props
```

```python
from pathlib import Path


polyinfo_dir = Path("../raw/PoLyInfo_extract_20250712")
polyinfo_files = list(polyinfo_dir.glob("*.csv"))

polyinfo_prop_list = []
polyinfo_desc_list = []

for file in polyinfo_files:
    polyinfo_data = pd.read_csv(file, index_col=0)
    polyinfo_prop_name = file.stem[3:].replace("_", " ")
    polyinfo_prop_list.append(
        polyinfo_data[[polyinfo_prop_name + " mean"]].rename(columns={polyinfo_prop_name + " mean": polyinfo_prop_name})
    )

    polyinfo_desc = polyinfo_data[FF_desc_names]
    polyinfo_desc_list.append(polyinfo_desc)

polyinfo_desc = pd.concat(polyinfo_desc_list, axis=0).loc[lambda df: ~df.index.duplicated(keep="first")]
polyinfo_prop = pd.concat(polyinfo_prop_list, axis=1).loc[polyinfo_desc.index]
```

```python
polyinfo_desc.to_parquet("../PolyInfo_FFDescriptor_20250712.parquet")
polyinfo_desc
```

```python
pipes = {}
scaled_props_list = []

# Process each property separately
for col in polyinfo_prop.columns:
    pipe = Pipeline(
        [("standardscaler", StandardScaler()), ("powertransformer", PowerTransformer(method="yeo-johnson"))]
    )
    # Fit and transform single column data
    data = polyinfo_prop[[col]].values
    scaled_data = pipe.fit_transform(data)
    # Store transformed data
    scaled_props_list.append(pd.DataFrame(data=scaled_data, index=polyinfo_prop.index, columns=[f"{col} (normalized)"]))
    # Store pipeline
    pipes[col] = pipe
# Save pipeline
joblib.dump(pipes, "../PolyInfo_properties_scaler_20250712.pkl.z")

# Combine all scaled properties
scaled_props = pd.concat(scaled_props_list, axis=1)

polyinfo_prop = pd.concat([polyinfo_prop, scaled_props], axis=1)
```

```python
polyinfo_prop.describe()
```

```python
from sklearn.model_selection import train_test_split

train_idx, val_idx, test_idx = [], [], []

# Get property names without (normalized)
prop_cols = [col for col in polyinfo_prop.columns if not col.endswith("(normalized)")]

# Sort properties by number of non-null values (ascending)
prop_counts = polyinfo_prop[prop_cols].notna().sum().sort_values()

# Initialize split labels
split = pd.Series(index=polyinfo_prop.index, dtype="object")

# Target ratios
train_ratio, val_ratio, test_ratio = 0.7, 0.15, 0.15

# Process each property from least to most populated
for prop_name in prop_counts.index:
    # Get indices with non-null values for this property
    prop_idx = polyinfo_prop[polyinfo_prop[prop_name].notna()].index.tolist()

    # Separate already labeled and unlabeled indices
    labeled_idx = [idx for idx in prop_idx if pd.notna(split.loc[idx])]
    unlabeled_idx = [idx for idx in prop_idx if pd.isna(split.loc[idx])]

    if len(unlabeled_idx) == 0:
        continue

    # Count current labels for this property
    current_train = sum(split.loc[labeled_idx] == "train") if labeled_idx else 0
    current_val = sum(split.loc[labeled_idx] == "val") if labeled_idx else 0
    current_test = sum(split.loc[labeled_idx] == "test") if labeled_idx else 0

    total_samples = len(prop_idx)
    remaining_samples = len(unlabeled_idx)

    # Calculate target counts
    target_train = int(total_samples * train_ratio)
    target_val = int(total_samples * val_ratio)
    target_test = total_samples - target_train - target_val

    # Calculate needed counts
    need_train = max(0, target_train - current_train)
    need_val = max(0, target_val - current_val)
    need_test = max(0, target_test - current_test)

    # Adjust if we don't have enough remaining samples
    total_needed = need_train + need_val + need_test
    if total_needed > remaining_samples:
        # Prioritize train and test
        if need_train + need_test <= remaining_samples:
            need_val = remaining_samples - need_train - need_test
        else:
            # Scale down proportionally, prioritizing train and test
            scale = remaining_samples / (need_train + need_test)
            need_train = int(need_train * scale)
            need_test = remaining_samples - need_train
            need_val = 0

    # Split unlabeled data
    if remaining_samples < 5:
        # Too few samples, assign all to train
        split.loc[unlabeled_idx] = "train"
    else:
        idx_to_assign = unlabeled_idx.copy()

        # Assign train
        if need_train > 0:
            idx_train = idx_to_assign[:need_train]
            split.loc[idx_train] = "train"
            idx_to_assign = idx_to_assign[need_train:]

        # Assign val
        if need_val > 0 and len(idx_to_assign) > 0:
            idx_val = idx_to_assign[:need_val]
            split.loc[idx_val] = "val"
            idx_to_assign = idx_to_assign[need_val:]

        # Assign test
        if need_test > 0 and len(idx_to_assign) > 0:
            idx_test = idx_to_assign[:need_test]
            split.loc[idx_test] = "test"
            idx_to_assign = idx_to_assign[need_test:]

        # Assign any remaining to train
        if len(idx_to_assign) > 0:
            split.loc[idx_to_assign] = "train"

polyinfo_prop["split"] = split
```

```python
polyinfo_prop.to_parquet("../PolyInfo_properties_20250712.parquet")
polyinfo_prop
```

```python
# Get property names without (normalized)
prop_cols = [col for col in polyinfo_prop.columns if not col.endswith("(normalized)") and col != "split"]

# Calculate split ratios for each property
split_ratios = []

for prop_name in prop_cols:
    # Get indices with non-null values for this property
    prop_data = polyinfo_prop[polyinfo_prop[prop_name].notna()]

    if len(prop_data) == 0:
        continue

    # Count samples in each split
    split_counts = prop_data["split"].value_counts()
    total = len(prop_data)

    # Calculate ratios
    train_ratio = split_counts.get("train", 0) / total
    val_ratio = split_counts.get("val", 0) / total
    test_ratio = split_counts.get("test", 0) / total

    split_ratios.append({
        "Property": prop_name,
        "Total Samples": total,
        "Train Count": split_counts.get("train", 0),
        "Val Count": split_counts.get("val", 0),
        "Test Count": split_counts.get("test", 0),
        "Train Ratio": f"{train_ratio:.2%}",
        "Val Ratio": f"{val_ratio:.2%}",
        "Test Ratio": f"{test_ratio:.2%}"
    })

# Create summary dataframe
split_summary = pd.DataFrame(split_ratios)
split_summary = split_summary.sort_values("Total Samples", ascending=False)
split_summary
```

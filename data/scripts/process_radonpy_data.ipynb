```python
import pandas as pd
import joblib
from sklearn.discriminant_analysis import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PowerTransformer
```

```python
radonpy_data = pd.read_csv("../amorphous_polymer_all_data_FFDescriptor_20250730.csv", index_col=0)
radonpy_data.head()
```

```python
FF_desc_names = radonpy_data.columns.tolist()[-190:]
FF_desc = radonpy_data[FF_desc_names]
FF_desc.to_parquet("../amorphous_polymer_FFDescriptor_20250730.parquet")
FF_desc
```

```python
prop_names_with_stats = radonpy_data.columns.tolist()[:-214]
print(len(prop_names_with_stats))
prop_names = [p for p in prop_names_with_stats if not p.endswith(('_median', '_std', '_min', '_max', '_count'))]
print(len(prop_names))

props = radonpy_data[prop_names]
# Create a dictionary to store pipelines for each property
pipes = {}
scaled_props_list = []

# Process each property separately
for col in props.columns:
    pipe = Pipeline([
        ("standardscaler", StandardScaler()),
        ("powertransformer", PowerTransformer(method="yeo-johnson"))
    ])
    # Fit and transform single column data
    data = props[[col]].values
    scaled_data = pipe.fit_transform(data)
    # Store transformed data
    scaled_props_list.append(
        pd.DataFrame(data=scaled_data, 
                    index=props.index, 
                    columns=[f"{col} (normalized)"])
    )
    # Store pipeline
    pipes[col] = pipe
# Save pipeline
joblib.dump(pipes, "../amorphous_polymer_properties_scaler_20250730.pkl.z")

# Combine all scaled properties
scaled_props = pd.concat(scaled_props_list, axis=1)

props = pd.concat([props, scaled_props], axis=1)
```

```python
from sklearn.model_selection import train_test_split

train_idx, val_idx, test_idx = [], [], []

for prefix in ["VL", "PI", "PG"]:
    data = props[props.index.str.startswith(prefix)]
    idx = data.index.tolist()
    n = len(idx)
    if n < 5:
        train_idx.extend(idx)
        continue
    idx_train, idx_temp = train_test_split(idx, test_size=0.3, random_state=42)
    idx_val, idx_test = train_test_split(idx_temp, test_size=0.5, random_state=42)
    train_idx.extend(idx_train)
    val_idx.extend(idx_val)
    test_idx.extend(idx_test)

split = pd.Series(index=props.index, dtype="object")
split.loc[train_idx] = "train"
split.loc[val_idx] = "val"
split.loc[test_idx] = "test"
props["split"] = split
```

```python
props.to_parquet("../amorphous_polymer_properties_20250730.parquet")
props
```

```python
PI_idx = [i for i in props.index if "PI" in i]
PI_props = props.loc[PI_idx].dropna(axis=1, how="all")
PI_props.to_parquet("../amorphous_polymer_PI_properties_20250730.parquet")
PI_props
```

```python
other_props = props.loc[~props.index.isin(PI_idx)].dropna(axis=1, how="all")
other_props.to_parquet("../amorphous_polymer_non_PI_properties_20250730.parquet")
other_props
```

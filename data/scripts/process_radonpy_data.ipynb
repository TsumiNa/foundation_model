```python
import pandas as pd
```

```python
radonpy_data = pd.read_csv("../amorphous_polymer_all_data_FFDescriptor_20250730.csv", index_col=0)
radonpy_data.head()
```

```python
FF_desc_names = radonpy_data.columns.tolist()[-190:]
FF_desc = radonpy_data[FF_desc_names]
FF_desc.to_parquet("../amorphous_polymer_FFDescriptor_20250730.parquet")
FF_desc
```

```python
prop_names_with_stats = radonpy_data.columns.tolist()[:-214]
print(len(prop_names_with_stats))
prop_names = [p for p in prop_names_with_stats if not p.endswith(('_median', '_std', '_min', '_max', '_count'))]
print(len(prop_names))

props = radonpy_data[prop_names]
props.to_parquet("../amorphous_polymer_properties_20250730.parquet")
props
```

```python
import joblib
from sklearn.discriminant_analysis import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PowerTransformer

pipe = Pipeline(
    [("standardscaler", StandardScaler()), ("powertransformer", PowerTransformer(method="yeo-johnson"))]
)
scaled_props = pd.DataFrame(data=pipe.fit_transform(props), index=props.index, columns=[f"{col} (normalized)" for col in props.columns])
scaled_props.to_parquet("../amorphous_polymer_properties_scaled_20250730.parquet")
joblib.dump(pipe, "../amorphous_polymer_properties_scaler_20250730.pkl.z")

scaled_props
```

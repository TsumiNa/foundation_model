```python
import pandas as pd

dataset = pd.read_pickle("../qc_ac_te_mp_dos_reformat_20250615.pd.xz")
dataset.head(3)
```

```python
from xenonpy.descriptor import Compositions

composition_calc = Compositions(
    featurizers='classic',
    n_jobs=20,
)
desc = composition_calc.fit_transform(dataset).dropna()
desc.to_pickle("../qc_ac_te_mp_dos_composition_desc_20250615.pd.xz")
desc.to_parquet("../qc_ac_te_mp_dos_composition_desc_20250615.pd.parquet")
desc
```

```python
desc = pd.read_pickle("../qc_ac_te_mp_dos_composition_desc_20250615.pd.xz")
desc.head(3)
```

```python
import joblib

from sklearn.discriminant_analysis import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PowerTransformer

pipe = Pipeline(
    [("standardscaler", StandardScaler()), ("powertransformer", PowerTransformer(method="yeo-johnson"))]
)
preprocessing_object = joblib.load("../preprocessing_objects_20250615.pkl.z")

desc_trans = pipe.fit_transform(desc)
desc_trans = pd.DataFrame(desc_trans, columns=desc.columns, index=desc.index)

desc_trans.to_pickle("../qc_ac_te_mp_dos_composition_desc_trans_20250615.pd.xz")
desc_trans.to_parquet("../qc_ac_te_mp_dos_composition_desc_trans_20250615.pd.parquet")

desc_trans
```

```python
missing_idx = dataset.index.difference(desc_trans.index)
missing_idx
```

```python
preprocessing_object["composition_desc_scaler"] = pipe
preprocessing_object["dropped_idx"] = missing_idx.tolist()
```

```python
joblib.dump(preprocessing_object, "../preprocessing_objects_20250615.pkl.z")
```
